```python
deliveries_df['batting_team'].unique()
```

'Royal Challengers Bengaluru' => Royal Challengers Bangalore'

'Rising Pune Supergiant'      => 'Rising Pune Supergiants'

```python
# Replace values in 'Batting team' column
deliveries_df['batting_team'] = deliveries_df['batting_team'].replace({
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
})

# Verify the replacements
print(deliveries_df['batting_team'].unique())
```

```python
deliveries_df['bowling_team'].unique()
```

```python
# Replace values in 'Bowling team'   column
deliveries_df['bowling_team'] = deliveries_df['bowling_team'].replace({
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
})

# Verify the replacements
print(deliveries_df['bowling_team'].unique())
```

```python
# Save the processed data
deliveries_df.to_csv('/content/deliveries_update_2008_2024.csv', index=False)
```
